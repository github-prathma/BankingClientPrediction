```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("bank-additional-full.csv", delimiter=";")
```

```python
df
```

```python
df.shape
```

```python
df.info()
```

```python
# all input variables
inputX = [x for x in df.columns if x != 'y']
inputX
```

```python
# segregate categorical and numerical variables
numericalVariables = df._get_numeric_data().columns
numericalVariables = list(set(numericalVariables))
print("Numerical Variables: ")
print(numericalVariables)
print()

categoricalVariables = list(set(inputX) - set(numericalVariables))
print("Categorical Variables: ")
print(categoricalVariables)
```

```python
for i, col in enumerate(df.columns):   
    if col in categoricalVariables:
        fig = plt.figure(figsize=(14.5,8.5))
        df[col+'_percent'] = df.groupby(['y', col])[col].transform(lambda x: len(x)*100) / len(df)
        sns.barplot(x=col, y=col+'_percent', hue='y', data=df).set(xlabel=col, ylabel='Percent(%)') 
        plt.xlabel(xlabel=col,size=14, color='#4f4e4e',labelpad=10,fontweight='bold')
        plt.ylabel(ylabel='Percent(%)',size=14, color='#4f4e4e', labelpad=10, fontweight='bold')
        plotTitle = "Percentage Distribution: Categories in "+ col.capitalize() + " with y (yes & no)"
        plt.title(plotTitle,size=18, fontweight='bold')
        plt.show()
```

```python
df.describe()
```

```python
# violimn plot visualization
for i, col in enumerate(df.columns):   
    if col in numericalVariables:
        fig = plt.figure(figsize=(9.5,5.5))
        sns.catplot(x='y', y=col, kind='violin', data=df)
        plt.xlabel(xlabel='y',size=14, color='#4f4e4e',labelpad=10,fontweight='bold')
        plt.ylabel(ylabel=col,size=14, color='#4f4e4e', labelpad=10, fontweight='bold')
        plotTitle = "Numerical Distribution: "+ col.capitalize() + " for y (yes & no)"
        plt.title(plotTitle,size=18, fontweight='bold')
        plt.show()

```

```python
# boxplot visualization             
for i, col in enumerate(df.columns):   
    if col in numericalVariables:
        fig = plt.figure(figsize=(9.5,5.5))
        sns.boxplot(x = 'y',y=col, data = df)
        plt.xlabel(xlabel='y',size=14, color='#4f4e4e',labelpad=10,fontweight='bold')
        plt.ylabel(ylabel=col,size=14, color='#4f4e4e', labelpad=10, fontweight='bold')
        plotTitle = "Numerical Distribution: "+ col.capitalize() + " for y (yes & no)"
        plt.title(plotTitle,size=18, fontweight='bold')
        plt.show()
```

### Binning

```python
df['age'].value_counts()
```

```python
#Quantile-based discretization function
#https://pbpython.com/pandas-qcut-cut.html
# age: max 98 min 17
df['age'] = pd.qcut(df['age'], q=5, labels=['1', '2', '3', '4','5'])
df['age']= df['age'].astype('int32')
df['age'].value_counts()
```

<b> pdays Binning in range , 0-4, 4-998, >998 </b>

```python
df['pdays'].value_counts()
```

```python
df['pdays'] = pd.cut(df['pdays'], [0, 4, 998, 1000], labels=[1,2,3])
```

```python
df['pdays'].value_counts()
```

```python
df['pdays'] = pd.to_numeric(df['pdays'], errors='coerce')
df = df.dropna(subset=['pdays'])
df['pdays'] = df['pdays'].astype(int, copy=False)
```

```python
df['pdays'].dtype
```

### Handling unknown values 

```python
df['job'].value_counts()
```

```python
df['loan'].value_counts()
```

Replacing unknown with most frequent value in column

```python
attributesWithUnknown = ['job', 'marital', 'education', 'default', 'housing', 'loan']
```

```python
for col in attributesWithUnknown:
    mostFrequent = df[col].mode()[0]
    df[col] = df[col].replace('unknown', mostFrequent)
```

```python
df['job'].value_counts()
```

### One-hot encoding of categorical variables 

```python
df = pd.get_dummies(df, prefix=categoricalVariables, columns=categoricalVariables)
df.head()
```
